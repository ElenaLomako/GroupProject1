```python
import pandas as pd
from pathlib import Path
```

```python
# Import the CSV file
sales_count_csv = Path("Data/SalesCount.csv")
df = pd.read_csv(sales_count_csv, encoding='latin-1')
df
```

```python
df.dtypes
```

```python
# Get the column headers as a list
columns = df.columns

# Loop through the column headers starting from the 6th column
for i in range(5, len(columns)):
    try:
        # Split the header by '/' and take the last part (year)
        parts = columns[i].split('/')
        
        # Ensure that the extracted year is four digits
        if len(parts[-1]) == 2:
            parts[-1] = '20' + parts[-1]  # Assuming the data is for years in the 2000s
        
        new_column_name = parts[-1]
        
        # Rename the column with the extracted year
        df.rename(columns={columns[i]: new_column_name}, inplace=True)
    except ValueError:
        # If the header does not contain '/', continue to the next column
        pass
    
df.head()
```

```python
# create a new dataframe with only state names and the 
sales_count_clean_df = df.iloc[:, 4:]
sales_count_clean_df
```

```python
# aggregate data by year
sales_agg_df = sales_count_clean_df.groupby(lambda x:x, axis=1).sum()
sales_agg_df 
```

```python
#drop the years we are not looking into
sales_agg_df = sales_agg_df.drop(columns=["2008", "2009", "2021", "2022", "2023"])

sales_agg_df.columns
```

```python
sales_agg_df
```

```python
#final data frame with all states
final_sales_df = sales_agg_df.groupby("StateName").sum() 
final_sales_df
```

```python
#Create a dataframe with only Washington, Utah, Texas, Illinois, Virginia states
selected_states = ["WA", "UT", "TX", "IL", "VA"]
select_states_sales_df = final_sales_df[final_sales_df.index.isin(selected_states)]
select_states_sales_df
```
