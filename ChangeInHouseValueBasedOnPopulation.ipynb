```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
#Import CSV file
home_values_state_level_csv = Path("Data/HomeValuesStateLevel.csv")
df = pd.read_csv(home_values_state_level_csv)
df.head
```

```python
# Get the column headers as a list
columns = df.columns

# Loop through the column headers starting from the 6th column
for i in range(5, len(columns)):
    try:
        # Split the header by '/' and take the last part (year)
        parts = columns[i].split('/')
        
        # Ensure that the extracted year is four digits
        if len(parts[-1]) == 2:
            parts[-1] = '20' + parts[-1]  # Assuming the data is for years in the 2000s
        
        new_column_name = parts[-1]
        
        # Rename the column with the extracted year
        df.rename(columns={columns[i]: new_column_name}, inplace=True)
    except ValueError:
        # If the header does not contain '/', continue to the next column
        pass
    
df.head()
```

```python
#Drop unnecessary columns for unused years
df_clean = df.drop(columns = ["RegionID", "SizeRank", "RegionType", "StateName", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2021", "2022", "2023"])
df_clean.head()
```

```python
#Aggregate data by year (sum the price of all houses sold)
house_value_agg_df = df_clean.groupby(lambda x:x, axis=1).sum()
house_value_agg_df.head()
```

```python
#Dividing by 12 (months) to get yearly average price of houses by state
house_value_agg_df['2010'] = house_value_agg_df['2010'].divide(12)
house_value_agg_df['2011'] = house_value_agg_df['2011'].divide(12)
house_value_agg_df['2012'] = house_value_agg_df['2012'].divide(12)
house_value_agg_df['2013'] = house_value_agg_df['2013'].divide(12)
house_value_agg_df['2014'] = house_value_agg_df['2014'].divide(12)
house_value_agg_df['2015'] = house_value_agg_df['2015'].divide(12)
house_value_agg_df['2016'] = house_value_agg_df['2016'].divide(12)
house_value_agg_df['2017'] = house_value_agg_df['2017'].divide(12)
house_value_agg_df['2018'] = house_value_agg_df['2018'].divide(12)
house_value_agg_df['2019'] = house_value_agg_df['2019'].divide(12)
house_value_agg_df['2020'] = house_value_agg_df['2020'].divide(12)
house_value_agg_df.head()
```

```python
#Selected the states being used in our analysis
select_states_avg_house_value_df = house_value_agg_df[house_value_agg_df["RegionName"].isin(["Texas", "Utah", "Washington", "Illinois", "Virginia"])]
select_states_avg_house_value_df

#Change State Name to State
select_states_avg_house_value_df = select_states_avg_house_value_df.rename(columns = {"RegionName": "State"})
select_states_avg_house_value_df     

#Set the index in the first column
select_states_avg_house_value_df.set_index("State")
```

```python
select_states_avg_house_value_df = select_states_avg_house_value_df.rename(columns={year: str(year) for year in range(2010, 2021)})
select_states_avg_house_value_df.columns
```

```python
avg_price = select_states_avg_house_value_df[[str(year) for year in range(2010, 2021)]]
price_years = avg_price.keys()
```

```python
price_years
```

```python
#Plot DC Annual Avg Price
dc_avg, = plt.plot(price_years, select_states_avg_house_value_df.loc['Washington',["2010","2011","2012","2013","2014","2015","2016","2017","2018",
                                                                                "2019","2020"]], marker="o",
                        color="purple",label=select_states_avg_house_value_df.loc['Washington',"State"])
#Plot Utah Annual Avg Price
ut_avg, = plt.plot(price_years, select_states_avg_house_value_df.loc['Utah',["2010","2011","2012","2013","2014","2015","2016","2017","2018",
                                                                                "2019","2020"]], marker="o",
                        color="green",label=select_states_avg_house_value_df.loc['Utah',"State"])
#Plot Illinois Annual Avg Price
il_avg, = plt.plot(price_years, select_states_avg_house_value_df.loc['Illinois',["2010","2011","2012","2013","2014","2015","2016","2017","2018",
                                                                                "2019","2020"]], marker="o",
                        color="blue",label=select_states_avg_house_value_df.loc['Illinois',"State"])
#Plot Texas Annual Avg Price
tx_avg, = plt.plot(price_years, select_states_avg_house_value_df.loc['Texas',["2010","2011","2012","2013","2014","2015","2016","2017","2018",
                                                                                "2019","2020"]], marker="o",
                        color="orange",label=select_states_avg_house_value_df.loc['Texas',"State"])
#Plot West Virginia Annual Avg Price
va_avg, = plt.plot(price_years, select_states_avg_house_value_df.loc['Virginia',["2010","2011","2012","2013","2014","2015","2016","2017","2018",
                                                                                "2019","2020"]], marker="o",
                        color="red",label=select_states_avg_house_value_df.loc['Virginia',"State"])
#make graph legable
plt.yticks(ticks=range(0, 220000, 130000), labels=range(0, 220000, 130000))
plt.tick_params(axis='both', which='major', labelsize=10)
plt.title("Average Price of House Over Time (Select States)")
plt.ylabel("Price in USD")
plt.xlabel("Years")
plt.legend(handles=[dc_avg, ut_avg, il_avg, tx_avg, va_avg])
plt.savefig('avg_house_price_change_graph.jpg', format='jpg')
plt.show()
```

```python
#Calculating year over year percent change in the average house price by selected states
select_states_avg_house_value_df["% Change 2010 to 2011"] = ((select_states_avg_house_value_df["2011"] - select_states_avg_house_value_df["2010"]) / select_states_avg_house_value_df["2010"] * 100) 
select_states_avg_house_value_df["% Change 2011 to 2012"] = ((select_states_avg_house_value_df["2012"] - select_states_avg_house_value_df["2011"]) / select_states_avg_house_value_df["2011"] * 100)
select_states_avg_house_value_df["% Change 2012 to 2013"] = ((select_states_avg_house_value_df["2013"] - select_states_avg_house_value_df["2012"]) / select_states_avg_house_value_df["2012"] * 100)
select_states_avg_house_value_df["% Change 2013 to 2014"] = ((select_states_avg_house_value_df["2014"] - select_states_avg_house_value_df["2013"]) / select_states_avg_house_value_df["2013"] * 100)
select_states_avg_house_value_df["% Change 2014 to 2015"] = ((select_states_avg_house_value_df["2015"] - select_states_avg_house_value_df["2014"]) / select_states_avg_house_value_df["2014"] * 100)
select_states_avg_house_value_df["% Change 2015 to 2016"] = ((select_states_avg_house_value_df["2016"] - select_states_avg_house_value_df["2015"]) / select_states_avg_house_value_df["2015"] * 100)
select_states_avg_house_value_df["% Change 2016 to 2017"] = ((select_states_avg_house_value_df["2017"] - select_states_avg_house_value_df["2016"]) / select_states_avg_house_value_df["2016"] * 100)
select_states_avg_house_value_df["% Change 2017 to 2018"] = ((select_states_avg_house_value_df["2018"] - select_states_avg_house_value_df["2017"]) / select_states_avg_house_value_df["2017"] * 100)
select_states_avg_house_value_df["% Change 2018 to 2019"] = ((select_states_avg_house_value_df["2019"] - select_states_avg_house_value_df["2018"]) / select_states_avg_house_value_df["2018"] * 100)
select_states_avg_house_value_df["% Change 2019 to 2020"] = ((select_states_avg_house_value_df["2020"] - select_states_avg_house_value_df["2019"]) / select_states_avg_house_value_df["2019"] * 100)

select_states_avg_house_value_df
```
