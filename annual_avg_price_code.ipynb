```python
import matplotlib.pyplot as plt
import requests
from scipy import stats
import json
import pandas as pd
import csv

from pathlib import Path
from pprint import pprint

```

```python
#pull excel file
annual_avg_xlsx = Path("Data/Clean_Avg_Price.xlsx")
annual_avg_df = pd.read_excel(annual_avg_xlsx)
annual_avg_df
```

```python
annual_avg_df.columns
```

```python
#change 'Unnamed: 0' to STNAME to work with population df
annual_avg_df = annual_avg_df.rename(columns={"State":"STNAME"})
```

```python
annual_avg_df
```

```python
#get selected states
selected_states = ["District of Columbia", "Utah", "Texas", "Illinois", "West Virginia"]
select_states_avg_df = annual_avg_df.groupby("STNAME").sum() 
select_states_avg_df
select_states_total_avg_df = select_states_avg_df[select_states_avg_df.index.isin(selected_states)]
select_states_total_avg_df.sort_values(["STNAME"], ascending=False)
```
